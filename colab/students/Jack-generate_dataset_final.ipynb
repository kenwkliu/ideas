```python
import plotly.graph_objects as go

import pandas as pd
from datetime import datetime

threshold = 0.005;

def assign_value(number):
      if number < -threshold:
        return -1
      elif number > threshold:
        return 1
      else:
        return 0

def assign_binary_value(number):
      if number <= 0:
        return 0
      else: 
        return 1

# Create a output file
data = {'image_name': [], 'return': []}
df_out = pd.DataFrame(data)

ticker = "2800"
df = pd.read_excel('D:\\Jack\\HKU_project\\mkt_data\\' + ticker + '.xlsx')

timespan=40
ret_timespan=10
last_idx=2460
# last_idx=4950
step=2

for idx in range(1230):
    # data is in descending order
    end=last_idx-idx*step
    start=end-timespan+1

    my_layout = go.Layout(
        autosize=True,
        width=100,
        height=100,
        margin=dict(l=0, r=0, t=0, b=0)
    )

    fig = go.Figure(data=[go.Candlestick(x=df.loc[start:end,'Date'],
                    open=df.loc[start:end,'Open'],
                    high=df.loc[start:end,'High'],
                    low=df.loc[start:end,'Low'],
                    close=df.loc[start:end,'Close'])],
                    layout=my_layout)

    fig.add_trace(
        go.Scatter(
            x=df.loc[start:end,'Date'],
            y=df.loc[start:end,'SMAVG (50)  on Close'],
            line=dict(color="yellow")
        ))

    fig.add_trace(
        go.Scatter(
            x=df.loc[start:end,'Date'],
            y=df.loc[start:end,'SMAVG (200)  on Close'],
            line=dict(color="violet")
        ))

    fig.update_layout(xaxis_rangeslider_visible=False)
    fig.update_layout(showlegend=False)
    fig.update_layout(yaxis={'visible': False, 'showticklabels': False})
    fig.update_layout(xaxis={'visible': False, 'showticklabels': False})
    
#     fig.show()
    filename=ticker+ "_" + df.loc[end,'Date'].to_pydatetime().strftime("%Y%m%d")+ ".jpg"
    fig.write_image("D:\\Jack\\HKU_project\\dataset\\"+filename)
    ret = df.loc[start-ret_timespan,'Close']/df.loc[start,'Close']-1
    
    print(df.loc[end,'Date'].to_pydatetime().strftime("%Y%m%d")+"  "+
          df.loc[start,'Date'].to_pydatetime().strftime("%Y%m%d")+"  "+
          df.loc[start-ret_timespan,'Date'].to_pydatetime().strftime("%Y%m%d")+"  "+
          str(ret))
    
    # df_out.iloc[idx,0]=filename
    # df_out.iloc[idx,1]=ret
    new_df = pd.DataFrame({'image_name': [filename], 'return': [assign_value(ret)]})
    df_out = pd.concat([df_out, new_df])
    
df_out.to_excel('D:\\Jack\\HKU_project\\dataset\\dataset.xlsx', index=False)
```
