```
 import numpy as np
 import pandas as pd
 #used to grab the stock prices, with yahoo
import pandas_datareader as web
 from datetime import datetime
 #to visualize the results
 import matplotlib.pyplot as plt
 import seaborn
 
 #select start date for correlation window as well as list of tickers
 start = datetime(2018, 1, 1)
 symbols_list = ['BTC-USD','ETH-USD']
```

```
#array to store prices
symbols=[]

#pull price using iex for each symbol in list defined above
for ticker in symbols_list: 
    r = web.DataReader(ticker, 'yahoo', start)
    # add a symbol column
    r['Symbol'] = ticker 
    symbols.append(r)

# concatenate into df
df = pd.concat(symbols)
df = df.reset_index()
df = df[['Date', 'Close', 'Symbol']]
df.head()
```

```
 df_pivot = df.pivot_table(index='Date',columns='Symbol',values='Close')
 df_pivot.head()
```

```
 corr_df = df_pivot.corr(method='pearson')  ——————————————————————————————————————————————————————————————————————————————————————————###################====================
 #reset symbol as index (rather than 0-X)
 corr_df.head().reset_index()
 corr_df.head()
```

```
 #generate plot
 mask = np.zeros_like(corr_df)
 seaborn.heatmap(corr_df, cmap='RdYlGn', vmax=1.0, vmin=-1.0 , mask = mask, linewidths=2.5)
 plt.yticks(rotation=0) 
 plt.xticks(rotation=90) 
 plt.show()
```

```
# Calculate the stock price ratios of Pair A/B and Pair B/A over the date period
# The price ratio is "adjusted close price" ratio between the 2 stocks

stocks = df_pivot[['BTC-USD','ETH-USD']]

stocks['BTC/ETH'] = stocks['BTC-USD'] / stocks['ETH-USD']
stocks['ETH/BTC'] = stocks['ETH-USD'] / stocks['BTC-USD']

stocks
```

```
# Calculate the average price ratio of the Pair over the period                            _————————————————————++++++++++++++++++++++——————————————————————————
# Trade if the current price ratio is significantly different
PX_RATIO_THRESHOLD = 0.05

# Calculate 10 days the average Buy PAIR_STOCK_A / Sell PAIR_STOCK_B ratio
stocks['avgPxRatio_BTC'] = stocks['BTC/ETH'].rolling(10).mean() ###########Calculating the 10-days mean ratio#############
print('Average BUY/SELL price ratio of BTC/ETH :', stocks['avgPxRatio_BTC'][-1])

stocks['execPxRatio_BTC'] = stocks['avgPxRatio_BTC'] * (1 - PX_RATIO_THRESHOLD)
print("Buy BTC-USD and Sell ETH-USD when the BUY/SELL price ratio is smaller than:", stocks['execPxRatio_BTC'][-1])

print(f'-'*100)
print("The ratio can be reached when relatively ... ")
print('BTC-USD',"goes down and become under-valued and should buy it")
print('ETH-USD', "goes up and become over-valued and should sell it")
```

```
# Plot the averge ratio vs the price ratio threshold
stocks[['BTC/ETH','avgPxRatio_BTC','execPxRatio_BTC']].loc['2020-01-01':'2020-12-31'].plot(figsize=(12,8))
```

```
# Calculate 10 days the average Buy PAIR_STOCK_B / Sell PAIR_STOCK_A ratio   ——————————————————————————————————————————————————————++++++++++++——————————————————————————————
stocks['avgPxRatio_ETH'] = stocks['ETH/BTC'].rolling(10).mean() ###########Calculating the 10-days mean ratio#############
print("Average BUY/SELL price ratio of",'ETH/BTC',":", stocks['avgPxRatio_ETH'][-1])

stocks['execPxRatio_ETH'] =stocks['avgPxRatio_ETH'] * (1 - PX_RATIO_THRESHOLD)
print("Buy", 'ETH-USD', "and Sell", 'BTC-USD', "when the BUY/SELL price ratio is smaller than:", stocks['execPxRatio_ETH'][-1])

print(f'-'*100)
print("The ratio can be reached when relatively ... ")
print('ETH-USD', "goes down and become under-valued and should buy it")
print('BTC-USD', "goes up and become over-valued and should sell it")
```

```
# Plot the averge ratio vs the price ratio threshold
stocks[['ETH/BTC','avgPxRatio_ETH','execPxRatio_ETH']].loc['2020-01-01':'2020-12-31'].plot(figsize=(12,8))
```

```
#Buy BTC sell ETH when BTC/ETH is smaller than the defined execPxRatio

stocks = stocks.dropna()

s1 = stocks['BTC/ETH'] > stocks['execPxRatio_BTC']
s2 = stocks['execPxRatio_BTC'] < stocks['BTC/ETH']
buyBTC_sellETH = stocks[~(s1 | s2.shift(1))]

print("\nPotential transaction date(s): \n")
print(buyBTC_sellETH[[ "BTC/ETH","execPxRatio_BTC","BTC-USD","ETH-USD"]].loc['2020-01-01':'2020-12-31'])
```

```
#Buy ETH sell BTC when ETH/BTC is smaller than the defined execPxRatio

stocks = stocks.dropna()

s3 = stocks['ETH/BTC'] > stocks['execPxRatio_ETH']
s4 = stocks['execPxRatio_ETH'] < stocks['ETH/BTC']
buyETH_sellBTC = stocks[~(s3 | s4.shift(1))]

print("\nPotential transaction date(s): \n")
print(buyETH_sellBTC[[ "ETH/BTC","execPxRatio_ETH","BTC-USD","ETH-USD"]].loc['2020-01-01':'2020-12-31'])
```

```
#Calculate all transactions gain and loss

#Buy BTC
buyBTC = round(buyBTC_sellETH["BTC-USD"].sum(),2)

#Buy ETH
buyETH = round(buyETH_sellBTC["ETH-USD"].sum(),2)

#Sell BTC
sellBTC = round(buyETH_sellBTC["BTC-USD"].sum(),2)

#Sell ETH
sellETH = round(buyBTC_sellETH["ETH-USD"].sum(),2)


print(buyBTC)
print(buyETH)
print(sellBTC)
print(sellETH)

Revenue = - buyBTC - buyETH + sellBTC + sellETH
print("\n Total gain is: $",Revenue)
```
