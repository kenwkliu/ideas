```python
#Prepare the dataset
import pandas as pd

def load_dataset(file_path):
    df = pd.read_excel(file_path)
    images = df.iloc[:, 0].tolist()
    labels = df.iloc[:, 1].tolist()
    return images, labels
```

```python
#Preprocess the images
from PIL import Image

def preprocess_image(image_path):
    image = Image.open(image_path).resize((100, 100))
    image = image.convert('RGB')
    image = np.array(image) / 255.0  # Normalize the pixel values
    return image
```

```python
#Split the dataset into train and test sets
from sklearn.model_selection import train_test_split

def split_dataset(images, labels):
    return train_test_split(images, labels, test_size=0.2, random_state=0)
```

```python
#Build and train the CNN model
import tensorflow as tf

def build_model():
    model = tf.keras.Sequential([
#         ##verion 1
#         tf.keras.layers.Conv2D(64, (5,5), activation='relu', input_shape=(100, 100, 3)),
#         tf.keras.layers.MaxPooling2D((2, 2)),
#         tf.keras.layers.Flatten(),
#         tf.keras.layers.Dense(64, activation='relu'),
#         tf.keras.layers.Dense(1)


#         ##verion 2
#         tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(100, 100, 3)),
# #         tf.keras.layers.MaxPooling2D((2, 2)),
#         tf.keras.layers.Conv2D(48, (3,3), activation='relu'),
#         tf.keras.layers.MaxPooling2D((2, 2)),
# #         tf.keras.layers.Dropout(0.2),
#         tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
# #         tf.keras.layers.Dropout(0.5)
#         tf.keras.layers.MaxPooling2D((2, 2)),
# #         tf.keras.layers.Conv2D(96, (3,3), activation='relu'),
# #         tf.keras.layers.MaxPooling2D((2, 2)),
# #         tf.keras.layers.Dropout(0.2),
#         tf.keras.layers.Flatten(),
# #         tf.keras.layers.Dense(128, activation='relu'),
#         tf.keras.layers.Dropout(0.5),
#         tf.keras.layers.Dense(2, activation='softmax')            
        

#         ##verion 3
        tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(100, 100, 3)),
        tf.keras.layers.MaxPooling2D((2, 2)),
#         tf.keras.layers.Conv2D(48, (3,3), activation='relu'),
#         tf.keras.layers.MaxPooling2D((2, 2)),
#         tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
        tf.keras.layers.MaxPooling2D((2, 2)),
#         tf.keras.layers.Conv2D(96, (3,3), activation='relu'),
#         tf.keras.layers.MaxPooling2D((2, 2)),
#         tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256),
#         tf.keras.layers.Dropout(0.2),
        tf.keras.layers.Dense(3, activation='softmax')       
    ])
    model.summary()
#     model.compile(optimizer='adam', loss='mean_squared_error')
    model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
    return model

def train_model(images, labels,test_images,test_labels):
    model = build_model()
#     print(type(images))
#     print(type(labels))
    labels_2 = np.array(labels, dtype="int")
#     print(type(labels_2))
    history = model.fit(images, labels_2, validation_data=(test_images, test_labels), epochs=32, batch_size=64)
    
    # Extract accuracy values from the history object
    train_accuracy = history.history['accuracy']
    test_accuracy = history.history['val_accuracy']

    # Plot the accuracy values
    epochs = range(1, len(train_accuracy) + 1)
    plt.plot(epochs, train_accuracy, 'b', label='Training Accuracy')
    plt.plot(epochs, test_accuracy, 'r', label='Test Accuracy')
    plt.title('Training and Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
    
    return model
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart
x_axis_labels = ['0', '1', '2']
y_axis_labels = ['0', '1', '2']
degrees = np.arange(1000, 10000, 1000).reshape(3, 3)

# Create a figure and axis
fig, ax = plt.subplots()

# Create a colormap for the degrees
cmap = plt.cm.get_cmap('Blues')

# Plot the blocks and add degree colors as text
for i in range(len(x_axis_labels)):
    for j in range(len(y_axis_labels)):
        x = i - 0.5
        y = j - 0.5
        degree = degrees[j][i]
        color = cmap(degree / 9000)  # Normalize the degree to the range [0, 1]
        ax.add_patch(plt.Rectangle((x, y), 1, 1, color=color))
        ax.text(x + 0.5, y + 0.5, str(degree), ha='center', va='center', color='white')

# Set the tick labels and limits
ax.set_xticks(range(len(x_axis_labels)))
ax.set_yticks(range(len(y_axis_labels)))
ax.set_xticklabels(x_axis_labels)
ax.set_yticklabels(y_axis_labels)
ax.set_xlim(-0.5, len(x_axis_labels) - 0.5)
ax.set_ylim(-0.5, len(y_axis_labels) - 0.5)

# Set the aspect ratio and adjust the limits
ax.set_aspect('equal')
ax.set_xlim(-0.5, len(x_axis_labels) - 0.5)
ax.set_ylim(-0.5, len(y_axis_labels) - 0.5)

# Set the axis labels
ax.set_xlabel('Predicted Label', ha='center')
ax.set_ylabel('True Label', ha='center')

# Add a colorbar
cbar = plt.colorbar(plt.cm.ScalarMappable(cmap=cmap))
cbar.set_label('Degree')

# Show the chart
plt.show()
```
