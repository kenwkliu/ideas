```
## ---------------------------------------------------------------------------
## FINA6289  Enhanced ML Circuit Breaker with Pure Unsupervised Anomaly Detection
## Author     : Law Chamber Lain
## Student ID : 20310803
## Date       : 13 Jul 2025
##
## This program detects unusual trading patterns (anomalies) in market data
## without relying on fixed thresholds. It uses multiple algorithms and
## combines their outputs to provide robust alerts, suitable for management
## review and operational deployment.
## ---------------------------------------------------------------------------

# Section 1: Environment Setup
# ----------------------------
# Install required packages in Colab (once per notebook session).
!pip install -q pandas numpy matplotlib scikit-learn tensorflow keras scipy

# Section 2: Imports and Configuration
# ------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from sklearn.ensemble import IsolationForest
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.covariance import EllipticEnvelope
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import precision_score, recall_score, accuracy_score
from scipy import stats
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings('ignore')

# Core parameters
TOTAL_TRADES     = 1000    # Number of simulated records per run
INTERVAL_MINUTES = 5       # Time step between trades
BASE_PRICE       = 100     # Starting price for synthetic data
NUM_RUNS         = 5       # Number of simulation runs for trend analysis

# Section 3: Data Generation
# --------------------------
# Generates realistic synthetic market data without hardcoded normal or outlier splits.
def generate_market_data(seed):
    np.random.seed(seed)
    start = datetime.combine(datetime.today().date(),
                             datetime.strptime("09:30","%H:%M").time())
    timestamps = [start + timedelta(minutes=i*INTERVAL_MINUTES) for i in range(TOTAL_TRADES)]

    # Combine multiple random processes to mimic real market moves
    trend       = np.random.normal(0,   0.001, TOTAL_TRADES)
    volatility  = np.random.gamma(2,    0.01,  TOTAL_TRADES)
    mean_rev    = np.random.normal(0,   0.005, TOTAL_TRADES)
    random_walk = np.random.normal(0,   0.01,  TOTAL_TRADES)
    shock_flag  = np.random.choice([0,1],TOTAL_TRADES, p=[0.95,0.05])
    shock_size  = np.random.normal(0, 0.1, TOTAL_TRADES) * shock_flag

    # Calculate combined price change and features
    change = (trend + mean_rev + random_walk + shock_size) * volatility
    log_ret = np.log1p(change)
    prices  = BASE_PRICE * np.exp(np.cumsum(log_ret))
    pct_chg = change * 100
    volume  = np.random.lognormal(10, 0.5, TOTAL_TRADES)

    # Feature engineering for anomaly detection
    rolling_vol = pd.Series(pct_chg).rolling(10, min_periods=2).std().fillna(0)
    rolling_ma  = pd.Series(pct_chg).rolling(10, min_periods=1).mean()
    momentum    = pd.Series(pct_chg).diff().fillna(0)

    return pd.DataFrame({
        'Datetime': timestamps,
        'Stock_Price': prices,
        'Pct_Change': pct_chg,
        'Volume': volume,
        'Price_Volatility': rolling_vol,
        'Returns_MA': rolling_ma,
        'Price_Change_Momentum': momentum
    })

# Section 4: Unsupervised Anomaly Detector Class
# ----------------------------------------------
# Encapsulates multiple detection methods and ensemble voting logic.
class UnsupervisedAnomalyDetector:
    def __init__(self, random_state=42):
        self.random_state = random_state
        self.scaler = StandardScaler()

    # 4.1 Build a simple autoencoder neural network
    def _build_autoencoder(self, dim):
        inp = Input(shape=(dim,))
        x = Dense(64, activation='relu')(inp)
        x = Dense(32, activation='relu')(x)
        x = Dense(16, activation='relu')(x)
        x = Dense(32, activation='relu')(x)
        x = Dense(64, activation='relu')(x)
        out = Dense(dim, activation='linear')(x)
        model = Model(inp, out)
        model.compile(optimizer='adam', loss='mse')
        return model

    # 4.2 Statistical methods
    def _z_score(self, arr, threshold=3):
        return (np.abs(stats.zscore(arr)) > threshold).astype(int)

    def _iqr(self, arr, multiplier=1.5):
        q1, q3 = np.percentile(arr, [25,75])
        iqr = q3 - q1
        lb, ub = q1 - multiplier*iqr, q3 + multiplier*iqr
        return ((arr < lb) | (arr > ub)).astype(int)

    # 4.3 Machine learning methods
    def _isolation_forest(self, X):
        clf = IsolationForest(contamination='auto', random_state=self.random_state)
        return (clf.fit_predict(X) == -1).astype(int)

    def _dbscan(self, X):
        return (DBSCAN(eps=0.5, min_samples=5).fit_predict(X) == -1).astype(int)

    def _lof(self, X):
        return (LocalOutlierFactor(n_neighbors=20, contamination='auto').fit_predict(X) == -1).astype(int)

    def _elliptic(self, X):
        clf = EllipticEnvelope(contamination=0.1, random_state=self.random_state)
        return (clf.fit_predict(X) == -1).astype(int)

    # 4.4 Autoencoder-based detection
    def _autoencoder(self, X, percentile=95):
        ae = self._build_autoencoder(X.shape[1])
        ae.fit(X, X, epochs=50, batch_size=32,
               validation_split=0.1,
               callbacks=[EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)],
               verbose=0)
        recon = ae.predict(X, verbose=0)
        errors = np.mean((X - recon)**2, axis=1)
        threshold = np.percentile(errors, percentile)
        return (errors > threshold).astype(int)

    # 4.5 Ensemble detection and voting
    def detect(self, df):
        features = df[['Pct_Change','Stock_Price','Volume',
                       'Price_Volatility','Returns_MA','Price_Change_Momentum']].values
        X_scaled = self.scaler.fit_transform(features)

        methods = {
            'z_score'            : self._z_score(df['Pct_Change']),
            'iqr'                : self._iqr(df['Pct_Change']),
            'isolation_forest'   : self._isolation_forest(X_scaled),
            'dbscan'             : self._dbscan(X_scaled),
            'lof'                : self._lof(X_scaled),
            'elliptic_envelope'  : self._elliptic(X_scaled),
            'autoencoder'        : self._autoencoder(X_scaled)
        }

        # Assemble results matrix
        matrix = np.column_stack(list(methods.values()))

        # Weighted voting: more weight to complex methods
        weights = np.array([0.1,0.1,0.15,0.15,0.15,0.15,0.2])
        df['weighted_anomaly'] = (np.average(matrix, axis=1, weights=weights) > 0.5).astype(int)

        # Expose individual method outputs (for audit)
        for name, result in methods.items():
            df[f'{name}_anomaly'] = result

        return df

# Section 5: Single-Run Visualization
# -----------------------------------
print(" Single Simulation with Anomaly Visualization")
df_single = generate_market_data(seed=0)
detector = UnsupervisedAnomalyDetector(random_state=0)
df_single = detector.detect(df_single)

plt.figure(figsize=(12, 6))
plt.plot(df_single['Datetime'], df_single['Stock_Price'], color='lightgray', label='Price')
plt.scatter(df_single[df_single['weighted_anomaly']==0]['Datetime'],
            df_single[df_single['weighted_anomaly']==0]['Stock_Price'],
            color='blue', s=10, label='Normal Trades')
plt.scatter(df_single[df_single['weighted_anomaly']==1]['Datetime'],
            df_single[df_single['weighted_anomaly']==1]['Stock_Price'],
            color='red',  s=20, label='Anomalous Trades')
plt.title('Section 5: Normal vs. Anomalous Trades (Single Run)')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Section 6: Multi-Run Performance Trends
# ---------------------------------------
def simulate_run(seed):
    df = generate_market_data(seed)
    df = UnsupervisedAnomalyDetector(seed).detect(df)
    # Define pseudo ground truth as extreme moves > 3
    threshold = df['Pct_Change'].std() * 3
    y_true = (np.abs(df['Pct_Change']) > threshold).astype(int)
    y_pred = df['weighted_anomaly']
    return (
        precision_score(y_true, y_pred),
        recall_score(y_true, y_pred),
        accuracy_score(y_true, y_pred)
    )

results = [simulate_run(s) for s in range(NUM_RUNS)]
precisions, recalls, accuracies = zip(*results)
runs = np.arange(1, NUM_RUNS + 1)

plt.figure(figsize=(10, 6))
plt.plot(runs, np.array(precisions)*100, marker='o', color='green',  label='Precision', linewidth=2)
plt.plot(runs, np.array(recalls)*100,    marker='s', color='orange', label='Recall',    linewidth=2)
plt.plot(runs, np.array(accuracies)*100, marker='d', color='red',    label='Accuracy',  linewidth=2)
plt.title('Section 6: Model Performance over 5 Simulation Runs')
plt.xlabel('Run Number')
plt.ylabel('Rate (%)')
plt.xticks(runs)
plt.ylim(0, 100)
plt.legend(frameon=False)
plt.grid(color='gray', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# Section 7: Technical Summary & Business Takeaways
# -------------------------------------------------
prec_min, prec_max, prec_avg = min(precisions), max(precisions), np.mean(precisions)
rec_min, rec_max, rec_avg = min(recalls),    max(recalls),    np.mean(recalls)
acc_min, acc_max, acc_avg = min(accuracies), max(accuracies), np.mean(accuracies)

print("\n Section 7: Technical Performance Summary")
print(f"- Precision: min {prec_min:.2%}, max {prec_max:.2%}, avg {prec_avg:.2%}")
print(f"- Recall:    min {rec_min:.2%}, max {rec_max:.2%}, avg {rec_avg:.2%}")
print(f"- Accuracy:  min {acc_min:.2%}, max {acc_max:.2%}, avg {acc_avg:.2%}")

print("\n Section 7: Management Commentary")
print("1. Consistent Precision (~{:.2%}) ensures reliable detection of true anomalies.".format(prec_avg))
print("2. Balanced Recall (~{:.2%}) indicates effective capture of extreme market moves.".format(rec_avg))
print("3. Overall Accuracy (~{:.2%}) demonstrates model stability across runs.".format(acc_avg))
print("4. Fine-tuning opportunities:")
print("    Adjust autoencoder depth or ensemble weights to further improve recall.")
print("    Incorporate real trading features (e.g., bid-ask spread) for production rollout.")
print("    Continuous retraining with live data will adapt the model to evolving market conditions.")
```
